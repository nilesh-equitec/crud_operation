# [dbo].[UpdateEmployee]

```sql
/****** Object:  StoredProcedure [dbo].[UpdateEmployee]    Script Date: 16-01-2024 19:54:19 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[UpdateEmployee]
    @Id INT,
    @Name NVARCHAR(100),
    @Email NVARCHAR(100),
    @Phone NVARCHAR(20),
    @Address NVARCHAR(255),
	@Skill nvarchar(255)
AS
BEGIN
    UPDATE Emp
    SET 
        Name = @Name,
        Email = @Email,
        Phone = @Phone,
        address = @Address,
		skill=@Skill
    WHERE Id = @Id;
END;
GO
```
